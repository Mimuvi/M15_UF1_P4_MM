```python
#Importar las 3 bibliotecas que vamos a utilizar que son:
#numpy, panda y matplotlib.
#post: Solo use las primeras línea si está usando un Jupyter Notebook 
#% matplotlib inline 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
#Ejecutamos las DataFrame 
weather = pd.read_csv("london2018.csv") 
#Muestra todo el contenido de la DF de london2018.csv.
print(weather)
```

```python
#Ejecuta la DF de londonweather.csv.
mydata=pd.read_csv("londonweather.csv")
```

```python
#---------------------Primer gráfico de Pandas----------------------------
#plt: recibe un conjunto de valores X e Y los cuales muestra en el plano
#definido por los ejes como puntos unidos por líneas.

#Traza un gráfico en el que se observa la temperatura máxima que está 
#definida como el eje X que se traza de manera horizontal y el eje Y 
#son los 12 meses del año que se traza de manera vertical.
weather.plot(y='Tmax', x='Month')
#Muestra el gráfico
plt.show()
```

```python
#---------------------------Gráficos de lineas-----------------------------
#pasar el tipo de trama en el parámetro tipo de esta manera:
weather.plot (kind = 'line', y = 'Tmax', x = 'Month')
#Se podría ejecuta sin la linea "plt.show ()" pero si no se utiliza se
#mostrará el gráfico junto con la siguiente linea:
#<matplotlib.axes._subplots.AxesSubplot at 0x7fe2fc2d7e80>
plt.show ()
```

```python
#plot.line:muestra una gráfica de linea a partir de los datos contenidos 
#en el DF.
weather.plot.line(y = 'Tmax', x = 'Month')
#Muestra el gráfico
plt.show ()
```

```python
#---------------------Gráficos de varias líneas---------------------
#En la siguiente linea se hace un gráfico de la temperatura mínima y la 
#máxima de los 12 meses del año 
weather.plot (y = ['Tmax', 'Tmin'], x = 'Month') 
#Muestra el resultado de los dos parametros.
plt.show ()
```

```python
#Creación de la nueva columna llamada Tmed que será el promedio de la
#temperatura minina y máxima que la suma de ello se dividirá por dos.
weather['Tmed'] = (weather['Tmax'] + weather['Tmin'])/2
```

```python
#Gráfico lineal que genere las temperatura minina y máxima y 
#el promedio de las temperaturas 
weather.plot (y = ['Tmax', 'Tmin', 'Tmed'], x = 'Month') 
#Muestra el resultado de los tres parametros.
plt.show ()
```

```python
#Gráfica de barras
#Tipo de gráfico kind = 'bar' que este caso es de barras, muestra por 
#pantalla los datos de lluvia y de temperatura que habían en 2018.
weather.plot (kind = 'bar', y = 'Rain', x = 'Month') 
plt.show ()
```

```python
# Es igual al gráfico anterior pero al usar kind='barh mostrar un gráfico
# lineal.
weather.plot(kind='barh' , y='Rain', x='Month')
plt.show()
```

```python
#Usando un gráfico de barras muestra las temperaturas mínimas y máximas que 
# hubo en el año 2018.
weather.plot (kind = 'bar', y = ['Tmax', 'Tmin'], x = 'Month') 
plt.show ()
```

```python
#Usando un gráfico de barras muestra las temperaturas mínimas, medias y
#máximas que hubo en el año 2018.
weather.plot (kind = 'bar', y = ['Tmax','Tmed', 'Tmin'], x = 'Month')
plt.show ()
```

```python
#------------------------Diagrama de dispersión----------------
#kind = 'scatter': Un diagrama dispersión traza una serie de puntos que corresponden a dos variables y nos permite 
#determinar si existe una relación entre ellas.

#En este diagrama disperso muestra el sol contra la lluvia los 12 puntos es valor de la lluvia "Rain" en el eje vertical 
#y el sol "sun" en el horizontal.
weather.plot (kind = 'scatter', x = 'Sun', y = 'Rain') 
plt.show ()
```

```python
#------------------------Gráficos circulares----------------
#kind = 'pie': Gráfico circular enseña algún valor se puede asociar con un 
#grupo o categoría.
#Parametro que utiliza 'Sun' que contiene los datos de luz de ese año.
#Muestra por pantalla la luz solar que hubo en cada mes en el año 2018
weather.plot (kind = 'pie', y = 'Sun') 
plt.show ()
```

```python
#En el gráfico anterior la leyenda es muy grande y no se diferencia
#los meses sino que lo muestra la luz de cada mes por etiquetas del 0 al 11.

#Para que sea entendible, lo primero que se pone es un index con los nombres
#de los meses y se quita la leyenda utilizando "legend=False".
weather.index=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug',
               'Sep','Oct','Nov','Dec']
weather.plot(kind='pie', y = 'Sun', legend=False)
plt.show()
```

```python
#------Gráficos estadísticos y detección de eventos inusuales.--------------
#Muestra un gráfico estadístico de la DF de londonweather con los 
# primeros cuatro años y las 48 filas.
more_weather = pd.read_csv("londonweather.csv")
print(more_weather[0:48])
```

```python
#Muestra la descripción de la columna "Rain".
print(more_weather.Rain.describe())
```

```python
#-----------------Diagramas de caja----------------------------------------
#plot.box: Diagrama de caja en donde los datos provienen de la parte 
#superior  e inferior del diagrama.Los cuartiles son los límites cuando 
#divide los datos en cuartos.El cuadro son los datos de lluvia.

#El primer cuartil (Q1) está en la marca del 25% de los puntos que estan por
#debajo del  (Q1) el 75%.

#El (Q2) es el que esta a la mitad,el 50% de los 2 puntos estan por encima 
#de Q2 y por debajo del 50%.

#El (Q3) esta el justo en el 75% y tambien el 25% de datos por encima del 
#Q3 y por debajo el 75%.

#En conclusión la caja es el 50%, la parte superior es el Q3 y la parte 
#inferior es el Q1, el Q2 es la linea horizontal que esta dentro del Q2.
more_weather.plot.box (y = 'Rain') 
plt.show ()
```

```python
#----------------------Histogramas-----------------------------------------
#kind = 'hist: El histograma divide los datos por rangos o contenedores que
#se representan por barras, el ancho de la barra es el valor y la altura que
#son el número de veces que se produce el valor en ese rango.
#Por defecto son 10 contadores.
#Muestra las precipitaciones de la lluvia
more_weather.plot (kind = 'hist', y = 'Rain') 
plt.show ()
```

```python
#En este histograma se establece el parametros de bins a lista de valores 
#de 8 por lo cual se limites nos dan 7 bins y cada uno de los contenedores 
# tendrán un rango de 25mm.
more_weather.plot (kind = 'hist', y = 'Rain',
                   bins = [0,25,50,75,100,125,150,175]) 
plt.show ()
```

```python
#Muestra un histograma de precipitaciones de tres contenedores que al tiene 
#un rango de 0 a 25 que es seco, de 25 a 75 que es normal y de 75 a 
#175 es húmedo.
more_weather.plot.hist (y = 'Rain', bins = [0,25,75,175]) 
plt.show ()
```

```python
#-----------------Múltiples gráficos---------------------------------------
#Muestra por pantalla 4 gráficos de linea en los que se establece que el
#X e Y son los siguientes datos "la temperatura mínima y máxima, la 
#lluvia y la luz solar", también se especifica un parámetro subplots que 
#es verdadero ya que por defecto es falso.

#layout el primer valor es es el numero de filas y el segundo es el número
#de figuras de cada fila.

#figsize: el primer valor es el ancho de la figura y el segundo es la altura.

weather.plot (y = ['Tmax', 'Tmin', 'Rain', 'Sun'], 
              subplots = True, layout = (2,2), figsize = (10,5)) 
plt.show ()
```

```python
#Muestra por pantalla 4 gráficos de barras con los mismo datos y valores 
#que los anteriores.
weather.plot (kind = 'bar', y = ['Tmax', 'Tmin', 'Rain', 'Sun'], 
              subplots = True, layout = (2,2), figsize = (10,5)) 
plt.show ()
```

```python
#Muestra por pantalla 4 gráficos circulares con los mismo datos que los 
#anteriores pero con los valores distintos de ancho y altura y sin leyenda.
weather.plot (kind = 'pie', y = ['Tmax', 'Tmin', 'Rain', 'Sun'], 
               subplots = True,legend = False, layout = (2,2), 
               figsize = (10 , 10)) 
plt.show ()
```

```python
#Muestra un gráfico circular del porcentaje de lluvia de los meses del año
#2018 y usando la función "savefig () " en la que se asigna un nombre y una
#extensión que este es un archivo .png se guarda en el directorio del
#programa en el se esta ejecutando.
weather.plot (kind = 'pie', y = 'Rain', legend = False) 
plt.show () 
plt.savefig ('pie.png')
```
